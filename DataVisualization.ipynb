# Data Visualization
This notebook is used to visualize the results obtained.
The library which is mainly used here is `matplotlib.pyplot` which allows to draw plots of the results.

```python
import matplotlib.pyplot as plt
```

## Data
Here we save vectors containing all the results. Since we don't have a massive number of results we keep track of all of them as variables.

```python
# Latent space is: 4, 8, 16, 32, 64
LS = [4, 8, 16, 32, 64]
MLP_res = [0.5711, 0.6826, 0.7012, 0.7045, 0.7022]
GMF_res = [0.6403, 0.6651, 0.6726, 0.6968, 0.6636]

# Number of neighbors is: 10, 20, 50, 100, 200
Neighbors = [10, 20, 50, 100, 200]
EuclideanCF_res = [0.168, 0.241, 0.342, 0.433, 0.397]
CosineCF_res = [0.374, 0.451, 0.458, 0.447, 0.428]
```

```python
fig, ax = plt.subplots()

ax.set_xscale('log', base=2)
ax.grid(b = True, color='0.95')
ax.plot(LS, MLP_res, 'o--')
ax.plot(LS, GMF_res, 'o--')
ax.legend(['MLP','GMF'])
ax.title.set_text('NN-based models')
ax.set_xlabel('Latent factors')
ax.set_ylabel('HR')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.grid(b = True, color='0.95')
ax.plot(Neighbors, EuclideanCF_res, 'ro--')
ax.plot(Neighbors, CosineCF_res, 'go--')
ax.legend(['EuclideanCF','CosineCF'])
ax.title.set_text('Neighbors-based CF')
ax.set_xlabel('Neighbors')
ax.set_ylabel('HR')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_xscale('log', base=2)
ax.grid(b = True, color='0.95')
ax.plot(LS, MLP_res, 'o--')
ax.plot(LS, GMF_res, 'o--')
ax.axhline(y=0.433, color='r', linestyle='--')
ax.axhline(y=0.458, color='g', linestyle='--')
ax.legend(['MLP','GMF','EucCF-100','CosCF-50'])
ax.title.set_text('Comparison')
ax.set_xlabel('Latent factors')
ax.set_ylabel('HR')

plt.show()
```
